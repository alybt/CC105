```python
print(df['familysize'].value_counts().sort_index())
```

```python
Q1 = df['familysize'].quantile(0.25)
Q3 = df['familysize'].quantile(0.75)
IQR = Q3 - Q1

upper_fence = Q3 + 1.5 * IQR
print(f"Statistically, anything above {upper_fence} is an outlier.")
```

```python
df = df[df['familysize'] < 30]

df['familysize_cleaned'] = df['familysize'].clip(upper=7.0)

print(df['familysize_cleaned'].value_counts().sort_index())
```

Some of the Majors are wrongs spellings, and those who have 2 I narrow it to one 

```python

architecture = [
    'Arsitechture', 'Archules', 'Architetecture ', 'Architecture/business', 'Architecture,planning and surveying ', 'Architecture,planning and surveying ', 
    'Architecture engineering ', 'Architecture', 'Architecture', 'Architectural technology', 
    'Architectural Design '
]

df['major'] = df['major'].str.lower().str.strip() 
df['major'] = df['major'].replace(architecture, 'Architecture')

print(df['major'].value_counts().get('architecture'))
```

<h6>Junk Responses</h6>

```python

junk_responses = [
    'never been', 'nil', 'no', 'no answer', 'none', 'none ', 
    'none yet', 'nope', 'not decide yet', 'not decided', 
    'not in university yet :)', 'not know', 'not sure yet', 'not yet'
]

df['major'] = df['major'].str.lower().str.strip() 
df['major'] = df['major'].replace(junk_responses, 'Undecided')

print(df['major'].value_counts().get('Undecided'))
```
